```python
#Q7#

# Importe as bibliotecas necessárias
from bs4 import BeautifulSoup
from scraper import scraper
from downloader import baixar_pagina

# Exemplo de uso da função scraper para uma URL de notícia
url = "https://bandnewstv.uol.com.br/conteudo/moraes-pede-informacoes-sobre-suposta-ameaca-a-lessa"
dados_noticia = scraper(url)

if dados_noticia:
    print("Dados da notícia raspados com sucesso:")
    for key, value in dados_noticia.items():
        print(f"{key}: {value}")
else:
    print("Falha ao raspar os dados da notícia.")

# Adicionar logs detalhados para depuração
print("Verificação dos elementos HTML:")
html_content = baixar_pagina(url)
soup = BeautifulSoup(html_content, 'html.parser')

print("Título:")
title_element = soup.find('h1', class_='main-title')
print(title_element.get_text(strip=True) if title_element else "Título não encontrado")

print("Data de publicação:")
datetime_element = soup.find('span', class_='data')
print(datetime_element.get_text(strip=True) if datetime_element else "Data de publicação não encontrada")

print("Autor:")
author_element = soup.find('p', class_='name')
print(author_element.get_text(strip=True) if author_element else "Autor não encontrado")

print("Texto:")
text_container = soup.find('div', class_='text')
if text_container:
    text_elements = text_container.find_all('p')
    text = " ".join([p.get_text(strip=True) for p in text_elements]) if text_elements else "Texto não encontrado"
else:
    text = "Contêiner de texto não encontrado"
print(text)
```

```python
#Q8#
import logging
import json

# Configurar o logger
logging.basicConfig(level=logging.DEBUG, format='%(asctime)s - %(levelname=s - %(message=s')
logger = logging.getLogger()

# Importe as funções do crawler, scraper e banco de dados
from crawler import coletar_todas_urls
from scraper import scraper
from database import criar_banco_de_dados, inserir_noticia, ler_banco_de_dados

# Caminho para o arquivo JSON de crawling
CRAWLING_INFO_PATH = '/content/drive/MyDrive/Colab Notebooks/crawler_crawling_info.json'

# Carregar o metadata de crawling
with open(CRAWLING_INFO_PATH, 'r') as file:
    metadata = json.load(file)

# Criar o banco de dados e a tabela de notícias
criar_banco_de_dados()

# Coletar URLs de notícias de todos os sites com paginação
num_paginas = 2  # Defina o número de páginas a serem coletadas
todas_urls = coletar_todas_urls(metadata, num_paginas)
logger.info(f"Total de URLs coletadas: {len(todas_urls)}")

# Raspagem de dados e inserção no banco de dados
for url in todas_urls:
    dados_noticia = scraper(url)
    if dados_noticia:
        logger.info(f"Dados da notícia raspados com sucesso para a URL: {url}")
        inserir_noticia(dados_noticia)
    else:
        logger.error(f"Falha ao raspar os dados da notícia para a URL: {url}")

# Ler o banco de dados e exibir as notícias
df_noticias = ler_banco_de_dados()
logger.info("Notícias no banco de dados:")
print(df_noticias)
```
