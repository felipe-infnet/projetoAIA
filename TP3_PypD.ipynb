```python
import pandas as pd

#URL GitHub
url = 'https://raw.githubusercontent.com/cassiusf/datasets/main/car_crashes.xlsx'

#carga do arquivo excel
xls_github = pd.ExcelFile(url)

# carga de cada planilha em um df
df_abbrev_github = pd.read_excel(xls_github, sheet_name='abbrev')
df_data_github = pd.read_excel(xls_github, sheet_name='data')
df_insurance_github = pd.read_excel(xls_github, sheet_name='insurance')

#primeiras linhas de cada dataframe
print("Conteúdo da planilha 'abbrev' (GitHub):")
df_abbrev_github.head()

print("Conteúdo da planilha 'data' (GitHub):")
df_data_github.head()

print("Conteúdo da planilha 'insurance' (GitHub):")
df_insurance_github.head()
```

```python
###Questao 2###
#agrupando tipo (ins_type)
grouped = df_insurance_github.groupby('ins_type')

#calculando as métricas para ins_premium
ins_premium_stats = grouped['ins_premium'].agg(['mean', 'max', 'min'])

#calculando as métricas para ins_losses
ins_losses_stats = grouped['ins_losses'].agg(['mean', 'sum', 'max', 'min'])

#resultados
print("Métricas de ins_premium agrupadas por ins_type:")
print(ins_premium_stats)

print("\nMétricas de ins_losses agrupadas por ins_type:")
print(ins_losses_stats)
```

```python
###Questao 3###
import pandas as pd
from sqlalchemy import create_engine, inspect

#path para o arquivo sql
db_path = 'mpg.db'

#conexão com o banco de dados SQLite
engine = create_engine(f'sqlite:///{db_path}')

#inspect nas tabelas no banco de dados
inspector = inspect(engine)
tables = inspector.get_table_names()

#saida das tabelas
print("Tabelas encontradas no banco de dados:")
print(tables)

#carga e exibicao das primeiras linhas de cada tabela
dataframes = {}
for table in tables:
    df = pd.read_sql_table(table, engine)
    dataframes[table] = df
    print(f"\nConteúdo da tabela '{table}':")
    print(df.head())
```

```python
###Questao 4###
from sqlalchemy import Column, Integer, String, MetaData, Table, create_engine
from sqlalchemy.orm import sessionmaker
from sqlalchemy.exc import SQLAlchemyError

db_path = 'mpg.db'  #path para o DB

#conector do banco de dados SQL
engine = create_engine(f'sqlite:///{db_path}')
metadata = MetaData()

#definir a nova tabela Category
category_table = Table('Category', metadata,
                       Column('category_id', Integer, primary_key=True),
                       Column('category', String))

#criar a tabela no banco de dados
try:
    metadata.create_all(engine)
    print("Tabela 'Category' criada com sucesso.")
except SQLAlchemyError as e:
    print(f"Erro ao criar a tabela: {e}")

#abrir uma sessao
Session = sessionmaker(bind=engine)
session = Session()

#inserir um registro na tabela criada
try:
    insert_query = category_table.insert().values(category_id=0, category='sedan')
    session.execute(insert_query)
    session.commit()
    print(f"Inserção realizada com sucesso.")
except SQLAlchemyError as e: #execpt porque estava com erros
    session.rollback()
    print(f"Erro ao inserir: {e}")
finally:
    session.close()

#verifica a tabela criada e o registro inserido
try:
    df_category = pd.read_sql_table('Category', engine)
    print("Conteudo da tabela 'Category':")
    print(df_category)
except SQLAlchemyError as e:
    print(f"Erro ao ler: {e}")
```

```python
###Questao 5###
import pandas as pd
from sqlalchemy import create_engine

db_path = 'mpg.db'

#conector do banco de dados SQL
engine = create_engine(f'sqlite:///{db_path}')

#carga das tabelas em df
df_category = pd.read_sql_table('Category', engine)
df_name = pd.read_sql_table('Name', engine)
df_observation = pd.read_sql_table('Observation', engine)
df_origin = pd.read_sql_table('Origin', engine)

#juncao das tabelas
df_merged = df_observation.merge(df_name, on='name_id', suffixes=('_observation', '_name'))
df_merged = df_merged.merge(df_origin, on='origin_id', suffixes=('', '_origin'))

#exibir o Df resultante
print("DataFrame Agregado:")
print(df_merged.head())

#Adicionado a categoria 'sedan' ao df final
df_merged['category'] = 'sedan'
print("\nDataFrame Agregado com Categoria:")
print(df_merged.head())
```

```python
###Questao 6###

#Para eliminar apenas precisamos executar o merge drop
df_cleaned = df_merged.drop(columns=['name_id', 'origin_id'])

#df resultante
print("DataFrame após eliminar 'name_id' e 'origin_id':")
print(df_cleaned.head())
```

```python
###Questa7###
#consumo médio (mpg) por pais
average_mpg_by_origin = df_cleaned.groupby('origin')['mpg'].mean()

# Determinar o país de origem com o menor consumo médio (mpg)
pais_maior_consumo = average_mpg_by_origin.idxmin()
menor_media_mpg = average_mpg_by_origin.min()

# Exibir o resultado
print("Consumo medio por país:")
print(average_mpg_by_origin)

print(f"\nPais de origem que mais consomem combustivel: {pais_maior_consumo} fazendo {menor_media_mpg} mpg")
```

```python
###Questao 8###
#groupby por pais e por quantidade de cilindros
grouped = df_cleaned.groupby(['cylinders', 'origin'])['weight'].mean()

#unstack para organizar na dimensao das colunas
average_weight = grouped.unstack()

#resultado
print("Peso médio dos carros por país e por quantidade de cilindros:")
print(average_weight)
```

```python
###Questa 9###
#groupby por ano de fabricação, quantidade de cilindros e calcular a media de HP
grouped_median = df_cleaned.groupby(['model_year', 'cylinders'])['horsepower'].median()

#unstack para organizar os cilindros nas colunas
median_horsepower = grouped_median.unstack()

#preencher os NAN com zero
median_horsepower = median_horsepower.fillna(0)

#resultado
print("Mediana da potência dos carros por ano de fabricação e por quantidade de cilindros:")
print(median_horsepower)
```

```python
###Questa10###
#groupby quantidade de cilindros e pais, e calcular as medias hp, aceleracao e peso
average_values = df_cleaned.groupby(['cylinders', 'origin'])[['horsepower', 'acceleration', 'weight']].mean()

#resultado
print("Médias da potência, aceleração e peso dos carros por quantidade de cilindros e país de origem:")
print(average_values)
```
